```python
import numpy as np
import pandas as pd
```

```python
Drugs = pd.read_excel("/content/drive/MyDrive/GDSC2_fitted_dose_response_25Feb20.xlsx")
Drugs.head()
```

```python
print(Drugs.DRUG_NAME.unique())
```

```python
from pubchempy import *
remove = []
identifiers = Drugs.DRUG_NAME.unique()
identifiers = identifiers.tolist()
ind = identifiers.index("Picolinici-acid")
identifiers[ind] = "Picolinic acid"
identifiers[identifiers.index("KRAS (G12C) Inhibitor-12")] = "K-Ras(G12C) inhibitor 12"
```

```python
for i in range(len(identifiers)):
    print(identifiers[i])
    try:
      identifiers[i] = get_properties(properties='canonical_smiles',identifier=identifiers[i].split(" (")[0], namespace='name')[0].get('CanonicalSMILES')
    except:
      remove.append(identifiers[i])
```

```python
print(len(remove))
```

```python
Drugs = Drugs[~Drugs.DRUG_NAME.isin(remove)]
```

```python
identifiers = [x for x in identifiers if x not in remove]
```

```python
names = Drugs.DRUG_NAME.unique()
print(len(names))
print(len(identifiers))
res = {names[i]: identifiers[i] for i in range(len(names))}
```

```python
Drugs = Drugs.replace({"DRUG_NAME": res})
```

```python
Drugs = Drugs.drop("NLME_RESULT_ID", axis = 1)
Drugs = Drugs.drop("DRUG_ID", axis = 1)
```

```python
Drugs.to_csv("Processed_Drug_Response.csv")
```
